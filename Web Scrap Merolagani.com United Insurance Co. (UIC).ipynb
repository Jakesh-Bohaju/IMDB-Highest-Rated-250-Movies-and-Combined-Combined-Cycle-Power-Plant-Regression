```python
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests
import csv
```

```python
url = requests.get("http://merolagani.com/CompanyDetail.aspx?symbol=BOKL")
soup = BeautifulSoup(url.content,'lxml')
table = soup.find_all('table',{'class':'table table-bordered table-striped table-hover'})

result=[]
for i in table:
    try:
        
        #date = i.find('td',{'class':'text-center'}).text
        #print(type(date))
        a_val = i.find_all('td')
        res1 = map(lambda x:x.text,a_val)
        res1 = list(res1)
        #print(res1)
        date = res1[1]
        ltp = res1[2]
        change = res1[3]
        high = res1[4]
        low = res1[5]
        ope = res1[6]
        qty = res1[7]
        turnover = res1[8]
        res2 = [date,ltp,change,high,low,ope,qty,turnover]
        
        result.append(res2)
    except Exception as e: 
        #print(e)
        pass
with open ('UIC.csv','w') as file:
   writer=csv.writer(file)
   writer.writerow(["Date", "LTP", "% Change", "High", "Low", "Open", "QTY", "Turnover"])
   for i in result:
     writer.writerow(i)
```
